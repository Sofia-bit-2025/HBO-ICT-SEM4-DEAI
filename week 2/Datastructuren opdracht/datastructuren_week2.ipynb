### Vraag 6

\[

b) \( g(n) = n^3 \)

c) \( g(n) = n^2 + n^3 \)

d) \( g(n) = n^5 \)

\[
n^2 + n^3 = O(n^3)

Elke student weet dit, op 
4 studenten na.

Hoeveel vragen moeten er maximaal worden gesteld voordat de docent het antwoord heeft?

### Antwoord

\[
\begin{aligned}

\end{aligned}
\]

### Vraag 12

In een klas zitten 
𝑛 studenten.

Een docent wil dat alle studenten elkaar leren kennen.
Hij noemt telkens twee studenten en laat hen zich aan elkaar voorstellen.

Dit doet hij voor alle mogelijke combinaties van twee studenten.

Hoe vaak moet de docent twee namen noemen voordat zijn doel is bereikt?

### Antwoord

\[
\begin{aligned}
\text{Aantal studenten} &= n \\
\text{Aantal studenten per ontmoeting} &= 2 \\
\\
\text{Aantal mogelijke paren van studenten} 
&= \binom{n}{2} \\
\\
&= \frac{n(n-1)}{2}
\end{aligned}
\]

### Vraag 26

\[
\begin{aligned}
&\text{Wat is de tijdscomplexiteit van deze methode?} \\
\\
&\texttt{for a in range}(0,n) \\
&\texttt{result = veryfst(result)} \quad (1\ \text{milliseconde per aanroep}) \\
\\
&\texttt{for b in range}(0,m) \\
&\texttt{result = veryslooooooooow(result)} \quad (1\ \text{uur per aanroep})
\end{aligned}
\]

### Antwoord

\[
\begin{aligned}
&\text{Eerste lus} &: n \text{ iteraties} \\
&\text{Kosten per iteratie} &: O(1) \\
&\text{Totale kosten} &: O(n) \\
\\
&\text{Tweede lus} &: m \text{ iteraties} \\
&\text{Kosten per iteratie} &: O(1) \\
&\text{Totale kosten} &: O(m) \\
\\
&\text{Totale tijdscomplexiteit} \\
&= O(n) + O(m) \\
&= O(n + m)
\end{aligned}
\]
